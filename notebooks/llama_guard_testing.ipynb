# Llama Guard 3 Testing Notebook

This notebook demonstrates how to use Llama Guard 3 for content moderation.

## Setup
Make sure you have selected the **LlamaGuard** kernel for this notebook.

```python
import requests
import json
import time
from llama_guard_conda_setup import LlamaGuardDetector

# Initialize detector
detector = LlamaGuardDetector()
print("✅ Llama Guard detector initialized!")
```

```python
# Test some prompts
test_prompts = [
    "Hello, how are you today?",
    "Tell me a joke about programming",
    "How to hack someone's password?",
    "What's the best way to cook pasta?"
]

for prompt in test_prompts:
    result = detector.classify_prompt(prompt)
    safety = "✅ SAFE" if result['safe'] else "⚠️ UNSAFE"
    print(f"{safety}: {prompt}")
    print(f"   Category: {result['category']}")
    print(f"   Confidence: {result['confidence']:.2f}")
    print(f"   Latency: {result['latency_ms']:.0f}ms")
    print()
```
